2. Definimos las funciones necesarias para hallar las órbitas asociadas a un Cuadrado Latino "L" y un autotopismo del mismo al que denominaremos "θ"

```python
#Conocemos que una órbita se define como aquel conjunto de entradas relativas a un CL y relacionadas entre sí de tal manera que,
#aplicando el autotopismo dado sobre una de dichas entradas, obtenemos otra que pertenece a la órbita en la que se encuentra la 
#primera.

def hallaOrbitas(cuadrado,autotopismo):
    #La idea principal es crear dos bucles anidados de tal manera que comprobemos cada uno de los elementos del cuadrado
    #evitando aquellos que ya hayan sido clasificados dentro de una órbita
    
    #########################
    # Variables principales #
    #########################
    
    visitados = []      #Almacena todos aquellos elementos que pertenecen a alguna órbita.
    orbita_actual = []  #Almacena la órbita actual con la que se está trabajando.
    orbitasCL = []      #Almacena el conjunto de órbitas presentes en el cuadrado latino dado.
    entrada_actual = (-1,-1,-1)  #Almacena la entrada del CL que estamos visitando en este momento.
    primera_entrada = (-1,-1,-1) #Almacena la que será la primera entrada de la orbita actual
    indice_actual = 0   #Alamcena el índice que estamos visitando dentro del CL
    
    ###############
    # Operaciones #
    ###############
    
    while(len(visitados)<len(cuadrado)): #Hasta que visitemos todos los elementos del cuadrado
        
        entrada_actual = cuadrado[indice_actual]
        orbita_actual = []
        
        if((entrada_actual not in visitados)):   #Comprobamos que la entrada que visitamos no se haya usado ya
            
            primera_entrada = entrada_actual   #Establecemos que la entrada inicial sea la actual
            orbita_actual.append(primera_entrada) #Incluimos la primera entrada en la orbita actual
            visitados.append(primera_entrada)  #Incluimos la primera entrada como visitado
            entrada_actual = aplicaIsotopismo(entrada_actual,autotopismo) #Aplicamos el autotopismo al elemento
            
            while(entrada_actual!=primera_entrada): #Bucle hasta que se cierre la órbita
                
                orbita_actual.append(entrada_actual)  #Incluimos la entrada en la orbita actual
                visitados.append(entrada_actual)      #Incluimos la entrada como visitado                
                entrada_actual = aplicaIsotopismo(entrada_actual,autotopismo) #Aplicamos el autotopismo al elemento
                                
        if(orbita_actual != []):
            orbitasCL.append(orbita_actual) #Incluimos la órbita obtenida en el conjunto de órbitas del CL

        indice_actual = indice_actual + 1 #Incrementamos el índice
        
    ##########
    # Salida #
    ##########
    
    return orbitasCL

#A modo de ayuda, definimos una funcion que haya las órbitas en un cuadrado latino parcial, partiendo de las entradas presentes
#en el mismo

def hallaOrbitasParcial(cuadrado, autotopismo):
    
    #########################
    # Variables principales #
    #########################
    
    visitados = []      #Almacena todos aquellos elementos que pertenecen a alguna órbita.
    orbita_actual = []  #Almacena la órbita actual con la que se está trabajando.
    orbitasCL = []      #Almacena el conjunto de órbitas presentes en el cuadrado latino dado.
    entrada_actual = (-1,-1,-1)   #Almacena la entrada del CL que estamos visitando en este momento.
    primera_entrada = (-1,-1,-1)  #Almacena la que será la primera entrada de la orbita actual
    indice_actual = 0             #Alamcena el índice que estamos visitando dentro del CL
    contadorEntradas = 0          #Almacena el numero de entradas no vacias
    contadorEntradasVisitadas = 0 #Almacena el numero de entradas no vacias visitadas
    
    ###############
    # Operaciones #
    ###############
    
    #Calculamos el número de entradas no vacías en el cuadrado
    for i in cuadrado:
        if(i!=[]):
            contadorEntradas = contadorEntradas + 1
            
    
    while((contadorEntradas!=contadorEntradasVisitadas) and (indice_actual!=len(cuadrado))): #Hasta que visitemos todos las entradas primarias del cuadrado
        
        entrada_actual = cuadrado[indice_actual]
        orbita_actual = []
        
        if((entrada_actual not in visitados) and entrada_actual!=[]):   #Comprobamos que la entrada que visitamos no se haya usado ya
            
            primera_entrada = entrada_actual   #Establecemos que la entrada inicial sea la actual
            orbita_actual.append(primera_entrada) #Incluimos la primera entrada en la orbita actual
            visitados.append(primera_entrada)  #Incluimos la primera entrada como visitado
            entrada_actual = aplicaIsotopismo(entrada_actual,autotopismo) #Aplicamos el autotopismo al elemento
            
            while(entrada_actual!=primera_entrada): #Bucle hasta que se cierre la órbita
                
                orbita_actual.append(entrada_actual)  #Incluimos la entrada en la orbita actual
                visitados.append(entrada_actual)      #Incluimos la entrada como visitado                
                entrada_actual = aplicaIsotopismo(entrada_actual,autotopismo) #Aplicamos el autotopismo al elemento
            
            contadorEntradasVisitadas = contadorEntradasVisitadas + 1 #Incrementamos el contador de entradas visitadas
                                
        if(orbita_actual != []):
            orbitasCL.append(orbita_actual) #Incluimos la órbita obtenida en el conjunto de órbitas del CL
        
        indice_actual = indice_actual + 1 #Incrementamos el índice
        
    ##########
    # Salida #
    ##########    
    return orbitasCL
```

```python
def aplicaIsotopismo(entrada,isotopismo):
    #El isotopismo se divide en tres partes: filas, columnas y símbolos, y realizan cambios sobre el CL
    #De la misma manera, cada entrada cuenta con una fila, una columna y un símbolo. 
    #Será objetivo de esta función la obtención de una nueva entrada a partir de una dada tras aplicarle un isotopismo
    #que se pasará como parámetro.
    
    #########################
    # Variables principales #
    #########################
    
    nueva_entrada = [-1,-1,-1] #Almacena la nueva entrada tras la aplicación del isotopismo sobre la entrada dada.
    indice_cambio = -1
    
    ###############
    # Operaciones #
    ###############
    
    for i in range(len(entrada)): #Recorremos los distintos elementos de la entrada (f,c,s).
        elemento = isotopismo[i]  #Tomamos la modificacion de fila, columna o simbolo pertinente.
        
        if(len(elemento)!=0):
            for e in elemento:
                if(entrada[i] in e):
                    indice_cambio = (e.index(entrada[i]) + 1) % len(e)
                    nueva_entrada[i] = e[indice_cambio]
                    break
                else:
                    nueva_entrada[i] = entrada[i]
        else:
            nueva_entrada[i] = entrada[i]
            
    ##########
    # Salida #
    ##########
    
    return nueva_entrada
```

3. Dado un cuadrado parcial y un autotopismo relativo al mismo, comprobamos si es únicamente completable 

```python
#Definimos una función que nos indique si, dado un cuadrado latino parcial cuyas entradas se suponen θ-conjunto critico de un CL
#y un autotopismo, dicho cuadrado será o no únicamente completable. De serlo, dichas entradas serán θ-conjunto crítico del cuadrado
#inicial

def esUnicamenteCompletable(cuadradoParcial, autotopismo):
    #La idea principal es tomar el cuadrado parcial, a cuyas entradas aplicaremos el autotopismo para obtener todas las entradas 
    #relativas a las órbitas a las que pertenecen. Tras este paso, se añadiran aquellas entradas que, forzosamente, no pueden tener otro
    #valor. Por último repetiremos estos pasos hasta que el tamaño del cuadrado latino no varíe.
    
    #########################
    # Variables principales #
    #########################
    
    ordenCuadrado = int(math.sqrt(len(cuadradoParcial)))
    cuentaEntradasVacias = cuadradoParcial.count([])    #Almacena cuantas entradas del cuadrado latino están vacias
    cuentaEntradasVaciasAux = 0                         #Almacena cuantas entradas del cuadrado latino están vacias - aux
    
    ###############
    # Operaciones #
    ###############
        
    #Igualamos las variables de la condicion de ruptura de bucle
    
    cuentaEntradasVaciasAux = cuentaEntradasVacias

    #Hallamos las órbitas de las entradas presentes y las incluimos en cuadrado Parcial
    orbitasCLP = hallaOrbitasParcial(cuadradoParcial,autotopismo)

    #Incluimos los nuevos elementos en el cuadrado Parcial
    for i in range(len(orbitasCLP)):
        for j in range(len(orbitasCLP[i])):
            entradaActual = orbitasCLP[i][j]
            if(entradaActual not in cuadradoParcial):
                lugarEntrada = entradaActual[0]*ordenCuadrado + entradaActual[1] #Calculamos la posicion que ocuparia el elemento
                cuadradoParcial[lugarEntrada] = orbitasCLP[i][j]
                    
    #Pintamos el cuadrado antes de completarlo forzosamente
    if(cuentaEntradasVacias!=0):
        #Comprobamos si existen elementos repetidos tras incluir las órbitas
        repetidos = compruebaElementosRepetidos(cuadradoParcial)

        if(repetidos):
            #En caso de ser True se corta la funcion y se retorna el valor
            #print("Existen valores repetidos en filas/columnas. No se puede considerar solución.")
            return (cuadradoParcial,False)

        #Una vez incluidas las órbitas, comprobamos qué entradas pueden ser completadas mediante técnicas triviales (forzoso y orbitas)
            
        cuadradoParcial, cuentaEntradasVacias, estado = completaTecnicasTriviales(cuadradoParcial, autotopismo)

    #Una vez se apliquen las técnicas de resolución triviales tendremos tres casos posibles:

    #1) El cuadrado latino no tiene solución posible / Tiene elementos repetidos
    if(estado == False):
        return (cuadradoParcial,False)
    
    #2) El cuadrado latino no tiene ninguna entrada vacía, por tanto se considera únicamente completable
    elif(cuentaEntradasVacias==0):
        
        return (cuadradoParcial,True)
    
    #3) Un cuadrado latino parcial al que tendremos que aplicar alguna técnica del estilo del backtracking para poder obtener
    #   las posibles colocaciones de las entradas (el resultado será únicamente / no únicamente completable)
    
    else:
        #aplicaBacktracking(cuadradoParcial)
        return backtrackingCL(cuadradoParcial, autotopismo, ordenCuadrado)
    
def completaTecnicasTriviales(cuadradoParcial, autotopismo):
    #La idea de esta función es que, a partir de un cuadrado latino parcial, se completen todas las entradas posibles de manera
    #forzosa. Para ello, haremos uso de un bucle similar al de la función anterior, de tal manera que para cada entrada vacía,
    #calcularemos cuales son los posibles valores que pueden incluirse en ella mediante operaciones de conjuntos.
    
    #########################
    # Variables principales #
    #########################
    
    ordenCuadrado = int(math.sqrt(len(cuadradoParcial)))
    cuentaEntradasVacias = cuadradoParcial.count([])    #Almacena cuantas entradas del cuadrado latino están vacias
    cuentaEntradasVaciasAux = 0                         #Almacena cuantas entradas del cuadrado latino están vacias - aux
    
    simbolosPosibles = [s+1 for s in range(ordenCuadrado)] #Calculamos todos los posibles símbolos
    simbolosAux = [] #Almacena los valores que NO está permitido usar
    
    estado = True #Informa si se ha producido alguna incompatibilidad
  
    ###############
    # Operaciones #
    ###############
    
    while((cuentaEntradasVacias!=cuentaEntradasVaciasAux)):
        cuentaEntradasVaciasAux = cuentaEntradasVacias
        
        for i in range(len(cuadradoParcial)): #Recorremos la lista de entradas
            if(cuadradoParcial[i] == []):
                
                #Deginimos la fila y la columna del elemento
                filaElemento = i/ordenCuadrado    #Calculamos la fila en la que se encontraría el elemento
                columnaElemento = i%ordenCuadrado #Calculamos la columna en la que se encontraría este elemento
                
                #Calculamos los posibles símbolos para cada posición vacía
                simbolosPosiblesAux = encuentraPosiblesValoresForzosos(cuadradoParcial, i, simbolosPosibles)
                
                #En el caso de que no podamos colocar ningún símbolo en una de las casillas, el cuadrado no tiene solución
                if(len(simbolosPosiblesAux)==0): 
                    #print("No existe solucón posible debido a una incompatibilidad")
                    estado = False
                    return (cuadradoParcial, cuentaEntradasVacias, estado)
                    
                #En caso de tener una única opción como simbolo para incluir en la entrada:
                elif(len(simbolosPosiblesAux)==1): 
                    cuadradoParcial[i] = [int(filaElemento), columnaElemento, simbolosPosiblesAux[0]]
                    
                    #Tras añadir el elemento, completamos con su órbita
                    cuadradoParcialCopia = cuadradoParcial.copy()
                    cuadradoParcial = anadeNuevaOrbita(cuadradoParcial, autotopismo, cuadradoParcial[i], ordenCuadrado)

                #En cualquier otro caso no tenemos la certeza de qué simbolo colocar, por lo que lo dejaremos en manos de backtracking
                # o la siguiente iteración de este bucle
                #else:
                    
                    #print("Más de una posibilidad para la entrada con pos: ",[int(filaElemento),columnaElemento], ". Posibles símbolos:",simbolosPosiblesAux)
            
        
        # Comprobamos si se ha repetido algún elemento tras todas las operaciones
        repetidos = compruebaElementosRepetidos(cuadradoParcial)
        
        if(repetidos):
            
            #print("Existen valores repetidos en filas/columnas. No se puede considerar solución.")
            estado = False
            return (cuadradoParcial, cuentaEntradasVacias, estado)
            
        
        #Recalculamos el número de entradas vacías del CL
        cuentaEntradasVacias = cuadradoParcial.count([])          
                        
    return (cuadradoParcial, cuentaEntradasVacias, estado)   

def anadeNuevaOrbita(cuadradoParcial, autotopismo, entrada, ordenCuadrado):
    
    #La idea principal de esta función radica en la obtención de la órbita relativa a una entrada y un autotopismo determinado.
    #Esta órbita se incluirá en el cuadrado parcial
    
    #########################
    # Variables principales #
    #########################
    
    entradaInicial = entrada     #Variable que representa la entrada inicial
    entradaAuxiliar = [-1,-1,-1] #Variable que representa una entrada auxiliar (para saber cuando termina una órbita)
    orbita = []                  #Variable que almacena las entradas de la órbita
    
    ###############
    # Operaciones #
    ###############
    
    #Aplicamos un primer isotopismo a la entrada inicial y lo incluimos en el cuadrado
    entradaAuxiliar = aplicaIsotopismo(entradaInicial,autotopismo)
    
    #Si la entrada obtenida no se encuentra en el cuadrado (es decir, no es una órbita trivial) la incluimos
    if(entradaAuxiliar not in cuadradoParcial):
        posicionEnCuadrado = entradaAuxiliar[0]*ordenCuadrado + entradaAuxiliar[1]
        cuadradoParcial[posicionEnCuadrado] = entradaAuxiliar        
    
    #Continuamos calculando la órbita
    while(entradaInicial != entradaAuxiliar):
        
        #Obtenemos el siguiente elemento de la órbita
        entradaAuxiliar = aplicaIsotopismo(entradaAuxiliar,autotopismo)
        
        #Si la entrada obtenida no se encuentra en el cuadrado (es decir, no es una órbita trivial) la incluimos
        if(entradaAuxiliar not in cuadradoParcial):
            posicionEnCuadrado = entradaAuxiliar[0]*ordenCuadrado + entradaAuxiliar[1]
            cuadradoParcial[posicionEnCuadrado] = entradaAuxiliar        
    
    return cuadradoParcial  

def compruebaElementosRepetidos(cuadrado):
    #Esta función determinará si existen incompatibilidades por valores repetidos dentro del CL
    
    #########################
    # Variables principales #
    #########################
    
    ordenCuadrado = int(math.sqrt(len(cuadrado)))
    simbolos = [s+1 for s in range(ordenCuadrado)]
    repetido = False
    
    ###############
    # Operaciones #
    ###############
    i = 0
    while (i<len(cuadrado)):
        valorInicialI = i
        fila = list()
        columna = list()
        
        for j in range(ordenCuadrado):
            posFila = (i+j) % ordenCuadrado
            posCol = (i + j*ordenCuadrado) % len(cuadrado)
          
            entradaFila = cuadrado[posFila]
            entradaColumna = cuadrado[posCol]
            
            if(entradaFila!=[]):
                fila.append(entradaFila[2])
            if(entradaColumna!=[]):
                columna.append(entradaColumna[2])
            
        #Comprobamos si se repiten
        for s in simbolos:
            contadorEnFila = fila.count(s)
            contadorEnColumna = columna.count(s)
                
            if(contadorEnFila > 1 or contadorEnColumna > 1):
                repetido = True
        
        i = i + ordenCuadrado + 1 #Para que tome los elementos de la diagonal
        
    
    return repetido
```

```python
def encuentraPosiblesValoresForzosos(cuadradoParcial, posicionEnCuadrado, simbolosPosibles):
    
    simbolosAux = [] #Limpiamos la lista auxiliar
    ordenCuadrado = int(math.sqrt(len(cuadradoParcial)))
    
    
    filaElemento = posicionEnCuadrado/ordenCuadrado    #Calculamos la fila en la que se encontraría el elemento
    columnaElemento = posicionEnCuadrado%ordenCuadrado #Calculamos la columna en la que se encontraría este elemento
                
    #Buscamos, para este elemento, cuales son los que se encuentran en su fila y columna y los descartamos
    for j in range(ordenCuadrado):
        for k in range(ordenCuadrado):
            if((j==int(filaElemento) and k!=columnaElemento) or (j!=int(filaElemento) and k==columnaElemento)):
                pos = j*ordenCuadrado + k
                entrada = cuadradoParcial[pos]
                if(entrada!=[]):
                    simbolosAux.append(entrada[2])
                                
    #Una vez tenemos todos los símbolos que No se pueden colocar en esa posición, restamos a los posibles los obtenidos
    simbolosPosiblesAux = restaLista(simbolosPosibles,simbolosAux)                
              
    return simbolosPosiblesAux
```

```python
pila = [] #La pila será una lista que incluirá cada uno de los sucesivos estados del backtracking

def backtrackingCL(cuadradoParcial, autotopismo, ordenCuadrado):
    
    #El funcionamiento de esta función se basa en la suposición de valores para casillas que, mediante tecnicas triviales, 
    #podrían tener más de un valor posible. Las suposiciones se van encadenando en la pila y se irán manteniendo o eliminando
    #en función de los resultados obtenidos.    
    
    #########################
    # Variables principales #
    #########################
    
    posicionEntrada  = cuadradoParcial.index([])      #Tomamos el primer elemento que se encuentre vacío
    posicionEntradai = int(posicionEntrada/ordenCuadrado) #Calculamos la posición en filas de la entrada seleccionada
    posicionEntradaj = posicionEntrada%ordenCuadrado #Calculamos la posicion en columnas de la entrada seleccionada
    entradaBT = [posicionEntradai, posicionEntradaj, -1] #Se trata del primer elemento sobre el que supondremos un valor
    
    simbolosPosibles = [s+1 for s in range(ordenCuadrado)] #Lista con los posibles símbolos del cuadrado
    
    soluciones = [] #Variable que guarda la solución/soluciones obtenidas
    contador = 0 #Variable destinada a contar el numero de soluciones que se obtienen
    
    ###############
    # Operaciones #
    ###############
    
    #Calculamos los posibles valores que podemos asginar a esa casilla
    simbolosPosiblesAux = encuentraPosiblesValoresForzosos(cuadradoParcial, posicionEntrada, simbolosPosibles)
    
    #Asignamos el primer valor posible a la entrada que hemos tomado
    entradaBT[2] = simbolosPosiblesAux[0]
    
    #Y lo eliminamos de la lista de valores posibles
    simbolosPosiblesAux = restaLista(simbolosPosiblesAux, [simbolosPosiblesAux[0]])
    
    #Limpiamos la pila
    if(len(pila)!=0):
        for _ in range(len(pila)):
            pila.pop(0)

    
    #Añadimos el estado Inicial a la pila
    estadoInicial = [entradaBT, simbolosPosiblesAux, cuadradoParcial, soluciones, contador]
    pila.append(estadoInicial)
    
    #Realizamos la llamada a la función secundaria de Backtracking (que será la recursiva)
    
    return backtrackingRecursivo(entradaBT, simbolosPosiblesAux, cuadradoParcial, soluciones, contador,autotopismo)    
```

```python
def backtrackingRecursivo(entradaBTPila, simbolosPosiblesAuxPila, cuadradoParcialPila, solucionesPila, contadorPila, autotopismo):
    
    #########################
    #      Pasos Previos    #
    #########################
    
    
    #Hacemos una copia del cuadrado:
    cuadradoParcialCopia = list(cuadradoParcialPila)
    
    #Debemos incluir la entrada pasada por parámetro al cuadrado
    ordenCuadrado =  int(math.sqrt(len(cuadradoParcialPila)))
    posicionEntradaPila = int(entradaBTPila[0]*ordenCuadrado + entradaBTPila[1])
    cuadradoParcialCopia[posicionEntradaPila] = entradaBTPila
    
    entradasVaciasActual = 0
    
    #Obtenemos la órbita relativa a la nueva entrada y la incluimos
    
    copiaAuxiliar = list(cuadradoParcialPila)
    
    cuadradoParcialCopia = anadeNuevaOrbita(cuadradoParcialCopia, autotopismo, entradaBTPila, ordenCuadrado)
        
    
    #Ahora debemos completar mediante técnicas triviales para la nueva entrada incluida
    
    cuadradoParcialCopia,entradasVaciasActual,estado = completaTecnicasTriviales(cuadradoParcialCopia, autotopismo)
    
    if(estado):
        #Si tras la primera suposición aún quedan casillas vacías:
        if(entradasVaciasActual != 0):

            #########################
            # Variables principales #
            #########################

            posicionEntrada  = cuadradoParcialCopia.index([])      #Tomamos el primer elemento que se encuentre vacío
            posicionEntradai = int(posicionEntrada/ordenCuadrado)#Calculamos la posición en filas de la entrada seleccionada
            posicionEntradaj = posicionEntrada%ordenCuadrado #Calculamos la posicion en columnas de la entrada seleccionada
            entradaBT = [posicionEntradai, posicionEntradaj, -1] #Se trata del primer elemento sobre el que supondremos un valor

            simbolosPosibles = [s+1 for s in range(ordenCuadrado)] #Lista con los posibles símbolos del cuadrado

            ###############
            # Operaciones #
            ###############


            #Calculamos los posibles valores que podemos asginar a esa casilla
            simbolosPosiblesAux = encuentraPosiblesValoresForzosos(cuadradoParcialCopia, posicionEntrada, simbolosPosibles)

            #Debemos comprobar si la consulta de valores posibles ha originado una incompatibilidad (no se obtienen posibles simbolos
            # para la casilla que estamos recorriendo)
            if(len(simbolosPosiblesAux)==0):

                #En caso de que no tengamos valores posibles para completar existen dos opciones:

                #Opción A: Que aún queden posibilidades en el elemento anterior de la pila (el que pasamos por parámetros)

                #Opción B: Que no existan posibilidades en el elemento anterior de la pila. Aquí debemos realizar una nueva
                #Comprobación doble

                #Opción B.1: Que la pila siga albergando elementos, por lo que tomamos otra posibilidad

                #Opción B.2: Que la pila no tenga más elementos. Entonces retornamos el cuadrado actual y V o F si está completo o no

                if(len(pila)==0): #B.2
                    #Debemos comprobar si el contador marca alguna solución
                    if(contadorPila == 1):
                        return (solucionesPila[0],True)
                    else:
                        return (cuadradoParcialPila,False)

                else: #B.1
                    #Declaramos una serie de variables para seguir la búsqueda de opciones
                    seguirBuscandoOpciones = True
                    opcionEncontrada = False
                    quedanElementos = True

                    while(seguirBuscandoOpciones and quedanElementos):
                        #Accedemos al último elemento de la pila
                        ultimoElementoPila = pila[-1]

                        #Comprobamos si en ese elemento podemos tomar otro valor para la entrada establecida
                        if(len(ultimoElementoPila[1])!=0): #A
                            #Cambiamos el valor del simbolo de la entrada tomada
                            opcionEncontrada = True
                            seguirBuscandoOpciones = False

                            #Entrada[simbolos] = valoresposibles[0]
                            ultimoElementoPila[0][2] = ultimoElementoPila[1][0]

                            #Eliminamos la posibilidad de esa lista
                            ultimoElementoPila[1] = restaLista(ultimoElementoPila[1],[ultimoElementoPila[1][0]])

                            return backtrackingRecursivo(pila[-1][0], pila[-1][1], pila[-1][2], solucionesPila, contadorPila, autotopismo)

                        else: #B
                            #En caso de que no tengamos más posibilidades en el ultimo elemento de la pila, lo eliminamos
                            pila.pop(-1)

                            if(len(pila)==0):
                                quedanElementos = False

                                #Comprobamos si hay alguna solución
                                if(contadorPila == 1):
                                    return (solucionesPila[0],True)
                                else:
                                    return (cuadradoParcialPila,False)

            elif(len(simbolosPosiblesAux)==1):

                #En caso de que solo haya un posible valor
                entradaBT[2] = simbolosPosiblesAux[0]

                #Eliminamos la posibilidad y llamamos a la función recursiva
                return backtrackingRecursivo(entradaBT, [], cuadradoParcialCopia, solucionesPila, contadorPila, autotopismo)


            else:

                #Asignamos el primer valor posible a la entrada que hemos tomado
                entradaBT[2] = simbolosPosiblesAux[0]

                #Y lo eliminamos de la lista de valores posibles
                simbolosPosiblesAux = restaLista(simbolosPosiblesAux, [simbolosPosiblesAux[0]])

                #Añadimos el estado Inicial a la pila
                estadoInicial = [entradaBT, simbolosPosiblesAux, cuadradoParcialCopia, solucionesPila, contadorPila]
                pila.append(estadoInicial)

                #Realizamos la llamada a la función secundaria de Backtracking (que será la recursiva)

                return backtrackingRecursivo(entradaBT, simbolosPosiblesAux, cuadradoParcialCopia, solucionesPila, contadorPila,autotopismo)


        #En el caso de que tras la suposición no queden entradas vacías:

        else:
            #Si no hay más entradas vacías quiere decir que el cuadrado está completo, es decir, se ha encontrado una solución.

            #Incrementamos el contador de soluciones
            contadorPila+=1

            #Comprobamos si existe más de una solución
            if(contadorPila>1):
                #pila = []
                return (solucionesPila[0], False) #Devolvemos la primera solución (por ejemplo, para seguir el estándar)

            else:            
                #En caso de no existir más soluciones, la añadimos a la lista
                solucionesPila.append(cuadradoParcialCopia)

                #Comprobamos si hay más elementos en la pila
                if(len(pila)==0):
                    #Si no hay más elementos y el contador es 1, quiere decir que existe una unica solución
                    #Comprobamos si hay alguna solución
                        if(contadorPila == 1):
                            return (solucionesPila[0],True)
                        else:
                            return (cuadradoParcialPila,False)

                else:
                    #Si tenemos más de una entrada, debemos comprobar si podemos realizar alguna modificación
                    seguirBuscandoOpciones = True
                    opcionEncontrada = False
                    quedanElementos = True

                    while(seguirBuscandoOpciones and quedanElementos):
                        #Accedemos al último elemento de la pila
                        ultimoElementoPila = pila[-1]

                        #Comprobamos si en ese elemento podemos tomar otro valor para la entrada establecida
                        if(len(ultimoElementoPila[1])!=0):
                            #Cambiamos el valor del simbolo de la entrada tomada
                            opcionEncontrada = True
                            seguirBuscandoOpciones = False

                            #Entrada[simbolos] = valoresposibles[0]
                            ultimoElementoPila[0][2] = ultimoElementoPila[1][0]

                            #Eliminamos la posibilidad de esa lista
                            ultimoElementoPila[1] = restaLista(ultimoElementoPila[1],[ultimoElementoPila[1][0]])

                            return backtrackingRecursivo(pila[-1][0], pila[-1][1], pila[-1][2], solucionesPila, contadorPila, autotopismo)

                        else:
                            #En caso de que no tengamos más posibilidades en el ultimo elemento de la pila, lo eliminamos
                            pila.pop(-1)

                            if(len(pila)==0):
                                quedanElementos = False

                                #Comprobamos si hay alguna solución
                                if(contadorPila == 1):
                                    return (solucionesPila[0],True)
                                else:
                                    return (cuadradoParcialPila,False)
    
    else:
        seguirBuscandoOpciones = True
        opcionEncontrada = False
        quedanElementos = True

        while(seguirBuscandoOpciones and quedanElementos):
            #Accedemos al último elemento de la pila
            ultimoElementoPila = pila[-1]

            #Comprobamos si en ese elemento podemos tomar otro valor para la entrada establecida
            if(len(ultimoElementoPila[1])!=0):
                #Cambiamos el valor del simbolo de la entrada tomada
                opcionEncontrada = True
                seguirBuscandoOpciones = False

                #Entrada[simbolos] = valoresposibles[0]
                ultimoElementoPila[0][2] = ultimoElementoPila[1][0]

                #Eliminamos la posibilidad de esa lista
                ultimoElementoPila[1] = restaLista(ultimoElementoPila[1],[ultimoElementoPila[1][0]])

                return backtrackingRecursivo(pila[-1][0], pila[-1][1], pila[-1][2], solucionesPila, contadorPila, autotopismo)

            else:
                #En caso de que no tengamos más posibilidades en el ultimo elemento de la pila, lo eliminamos
                pila.pop(-1)

                if(len(pila)==0):
                    quedanElementos = False

                    #Comprobamos si hay alguna solución
                    if(contadorPila == 1):
                        return (solucionesPila[0],True)
                    else:
                        return (cuadradoParcialPila,False)
        
```

5. Calculamos el autotopismo θ' para el nuevo cuadrado L' (mediante el transformado por el isotopismo θ₁ de θ: θ₁* θ * θ₁⁻¹ )

```python
def autotopismoTransformado(autotopismo, isotopismoAleatorio, ordenCuadrado):
    #En esta función partiremos desde el isotopismo aleatorio al que aplicaremos el autotopismo inicial tomado y seguidamente
    #el inverso del isotopismo aleatorio. Así:
    
    #########################
    # Variables principales #
    #########################
    
    isotopismoInverso = [[],[],[]] #Variable que almacenará el inverso del isotopismo aleatorio
    entradasAuxiliares = []        #Variable que almacenará las entradas auxiliares para calcular el transformado
    
    #Variable que almacenará la relación entre los isotopismos para calcular el transformado
    transformado = {"filas":dict(),"columnas":dict(),"simbolos":dict()}
    
    ###############
    # Operaciones #
    ###############
    
    #En primer lugar calculamos el isotopismo inverso del isotopismo aleatorio
    
    for fcs in range(len(isotopismoAleatorio)):
        for camb in (isotopismoAleatorio[fcs]):
            isotopismoInverso[fcs].append(camb[::-1])
            
    print("El inverso del Isotopismo Aleatorio es:", isotopismoInverso)
    
    #Una vez tenemos calculado el inverso del isotopismo aleatorio, calculamos el autotopismo θ'
    
    #Para ello aplicaremos los tres autotopismos a unas entradas auxiliares que crearemos en función del orden del cuadrado
    
    for i in range(ordenCuadrado):
        #Entrada auxiliar previamente mencionada
        entradaOriginal = [i,i,i+1]
        
        #Realizamos el transformado
        
        #Aplicamos primero el isotopismo Inverso
        entradaTransformada = aplicaIsotopismo(entradaOriginal, isotopismoInverso)
        
        #Aplicamos el autotopismo
        entradaTransformada = aplicaIsotopismo(entradaTransformada, autotopismo)
        
        #Aplicamos el isotopismo aleatorio
        entradaTransformada = aplicaIsotopismo(entradaTransformada, isotopismoAleatorio)
        
        #Ahora incluimos en el diccionario definido al inicio de la función, la relación entre la entrada original y la
        #transformada
        
        transformado["filas"][i]=entradaTransformada[0]
        transformado["columnas"][i]=entradaTransformada[1]
        transformado["simbolos"][i+1]=entradaTransformada[2]
        
        
        
    #pintamos el transformado
    print("El transformado en diccionario es:", transformado)
    #Una vez tenemos la relación, procedemos a crear el autotopismo con la estructura que tenía habitualmente
    autotopismo_transformado = [[],[],[]]
    for i in range(len(transformado)):
        
        c = list(transformado.keys())
        cActual = c[i]
        valor = transformado[cActual]
        autoAux = []
        
        claves = list(valor.keys())
        
        primeraClave = claves[0]
        primerValor = valor[primeraClave]
        
        if(primeraClave == primerValor):
            del valor[primeraClave]
        else:
            autoAux.append(primeraClave)
            autoAux.append(primerValor)

            del valor[primeraClave]
            
            while(len(valor)!=0):
                siguienteValor = autoAux[-1]
                
                if(valor[siguienteValor] == autoAux[0]):

                    del valor[siguienteValor]
                    if(len(autoAux)!=0):
                        
                        if(len(autoAux)!=1):
                            autotopismo_transformado[i].append(autoAux)
                        autoAux = []
                        clavesAux = list(valor.keys())
                        
                        if(len(clavesAux)!=0):
                            aux = clavesAux[0]
                            autoAux.append(aux)
                    else:
                        autoAux = []
                        clavesAux = valor.keys()
                        if(len(clavesAux)!=0):
                            aux = clavesAux[0]
                            autoAux.append(aux)
                        
                else:
                    autoAux.append(valor[siguienteValor])
                    del valor[siguienteValor]
    ##########
    # Salida #
    ##########
        
    return autotopismo_transformado
```

```python
def estudioEstadistico(cuadrado, autotopismo, ordenConjunto):
    #La idea principal de esta función consiste en comprobar un número X de veces si existen conjuntos críticos de cierto Orden
    #para el cuadrado latino y autotopismo dados. 
    
    x = 1000          # Número de veces que se repetirá la prueba
    contadorAcierto = 0 # Variable que guarda el número de veces que encuentra un conjunto crítico
    contadorFallo   = 0 # Variable que guarda el número de veces que NO encuentra un conjunto crítico
    
    orbitasCuadrado = hallaOrbitas(cuadrado, autotopismo) #Calculamos las órbitas del cuadrado
    #print("Orbitas del cuadrado: ", orbitasCuadrado)
    combinacionesOrbitas = list(itertools.combinations(orbitasCuadrado, ordenConjunto))
    #print("combinaciones Obtenidas:",combinacionesOrbitas)
    ordenCuadrado = int(math.sqrt(len(cuadrado)))
    
    conjuntosCriticosObtenidos = list() #Almacenará los conjuntos críticos obtenidos
    
    #Ejecutamos el bucle X veces: 
    for i in range(x):

        #Seleccionaremos una combinación aleatoria entre las posibles y tomamos de cada una el primer elemento para formar 
        #un conjunto de entradas a evaluar
        combAleatoria = random.randint(0,len(combinacionesOrbitas)-1)
        combSeleccionada = combinacionesOrbitas[combAleatoria]
        
        #Formamos el conjunto a evaluar:
        
        conjunto = [a[0] for a in combSeleccionada]   
        
        #El siguiente paso es comprobar si es únicamente completable, por lo que, en primer lugar, formaremos un cuadrado latino 
        #parcial con las entradas obtenidas en el paso anterior
        cuadradoParcial = []
        
        #Creamos un cuadrado vacío inicial cuyo tamaño sea idéntico al del representante pasado por parámetro
        for _ in range(len(cuadrado)):
            cuadradoParcial.append([])
        
        #Para cada elemento del conjunto seleccionado, calculamos su posición y lo incluimos en el cuadrado parcial
        for j in conjunto:
            posicionEnCuadrado = j[0]*ordenCuadrado + j[1] 
            cuadradoParcial[posicionEnCuadrado] = j
            
        #Limpiamos la pila de backtraking para eliminar residuos
        
        #Comprobamos si es únicamente completable
        cuadradoAux, res = esUnicamenteCompletable(cuadradoParcial,autotopismo)
        
        if(res):
            #En caso de que sea únicamente completable, debemos comprobar para cada combinación de n-1 elementos del conjunto si 
            #es únicamente completable
            unicamenteCompletableSubconjunto = False #Variable booleana para comprobar cada subconjunto
            
            #Obtenemos todas las combinaciones de n-1 elementos para el conjunto inicial (suponiendo n el tamaño del conjunto inicial)
            tamSubconjunto = len(conjunto)-1
            subconjuntos = list(itertools.combinations(conjunto,tamSubconjunto))
            
            #Hacemos la comprobación con cada subconjunto
            for subconjunto in subconjuntos:

                res = compruebaSubconjunto(ordenCuadrado,subconjunto,autotopismo)
                #print("resultado Funcion:",res)
                
                unicamenteCompletableSubconjunto = unicamenteCompletableSubconjunto or res
            
                if(res):                    
                    break #Impedimos que se ejecute más
            
            if(unicamenteCompletableSubconjunto):
                #En caso de ser cierto, quiere decir que existe un subconjunto con un tamaño menor con el que también obtenemos el
                #cuadrado completo de manera única, por tanto incrementamos el contadorFallo
                contadorFallo += 1
                
            else: 
                #En caso contrario, quiere decir que no existen subconjuntos de tamaño menor para los que el cuadrado es únicamente
                #completable, por tanto el conjunto inicial tomado es crítico. Incrementamos entonces el contadorAciertos
                contadorAcierto += 1
                conjuntosCriticosObtenidos.append(conjunto)                
        
        else:
            #En caso de no ser unicamente completable incrementamos el valor del contadorFallo
            contadorFallo += 1
      
    #Test comprobacion
    listaResultados = list()
    
    for c in conjuntosCriticosObtenidos:
        cuadradoParcialAux = []
        
        #Creamos un cuadrado vacío inicial cuyo tamaño sea idéntico al del representante pasado por parámetro
        for _ in range(len(cuadrado)):
            cuadradoParcialAux.append([])
        
        #Para cada elemento del conjunto seleccionado, calculamos su posición y lo incluimos en el cuadrado parcial
        for j in c:
            posicionEnCuadradoAux = j[0]*ordenCuadrado + j[1] 
            cuadradoParcialAux[posicionEnCuadradoAux] = j
        
        cuadrado, res = esUnicamenteCompletable(cuadradoParcialAux, autotopismo)
        listaResultados.append(res)
        
    return (contadorAcierto, contadorFallo) 
```

```python
def compruebaSubconjunto(ordenCuadrado, entradas,autotopismo):
    #La idea principal de esta función es crear un cuadrado parcial a partir de las entradas que se pasan por parámetro y 
    #comprobar si es únicamente completable
    
    cuadradoParcial = [] #Variable que almacenará el cuadrado parcial
    numeroTotalEntradas = ordenCuadrado * ordenCuadrado
    
    #Rellenamos el cuadrado parcial con entradas vacías
    for _ in range(numeroTotalEntradas):
        cuadradoParcial.append([])
    
    #Incluimos las entradas que se pasan por parámetro al cuadrado parcial
    for i in entradas:
        posicionEnCuadrado = i[0]*ordenCuadrado + i[1] 
        cuadradoParcial[posicionEnCuadrado] = i
    
    #Comprobamos si es únicamente completable
    cuadradoAux, res = esUnicamenteCompletable(cuadradoParcial,autotopismo)
    
    return res
```

### Compartición de secretos al mismo nivel

Hasta ahora tenemos todas las herramientas necesarias para llevar a cabo una compartición de secretos cuyos participantes
se encuentren todos al mismo nivel de permisos. La idea principal radica en los siguientes puntos:

    1) Tomamos un cuadrado latino representante de una clase (esto variará en función de la cantidad de participantes, por ejemplo, para 3 participantes debemos buscar un cuadrado latino que, al aplicar uno de sus autotopismos, pueda encontrar un conjunto crítico de tamaño 3. Para ello, usar tabla de autotopismos del artículo).

    2) Encontrar un θ conjunto crítico del tamaño requerido

    3) Aplicar un isotopismo aleatorio θ1 al cuadrado L para obtener otro cuadrado L' isotópico al primero.

    4) Calcular el autotopismo θ' para el nuevo cuadrado L'

    5) Recuperar los θ'-conjuntos críticos para L' a partir de los θ-conjuntos críticos de L previamente encontrados

El secreto para el protocolo de comparticion de secretos sera el cuadrado L', mientras que las sombras a repartir entre los participantes se obtendran  entre los elementos de los θ′- conjuntos críticos.

```python
#El primer paso es crear un diccionario cuyas claves sean pares Representante - Autotopismo y los valores sean pares con sus
# respectivos scs y lcs asociados. Así:

comparteSecretos = {
    "L2_1" : [1,1],
    "L3_1" : [1,2],
    "L3_2" : [2,2],
    "L3_3" : [1,1],
    "L41_1" : [4,4],
    "L41_2" : [4,4],
    "L41_3" : [3,3],
    "L41_4" : [2,2],
    "L41_5" : [2,2],
    "L42_1" : [4,4],
    "L42_2" : [3,3],
    "L42_3" : [4,4],
    "L42_4" : [2,2],
    "L42_5" : [3,3],
    "L51_1" : [3,5],
    "L51_2" : [3,3],
    "L51_3" : [4,4],
    "L51_4" : [2,2],
    "L52_1" : [3,5],
    "L52_2" : [5,6]
    
}

relacion = {
    "L2_1" : (representante_L2, autotopismos_L2[1]),
    "L3_1" : (representante_L3, autotopismos_L3[1]),
    "L3_2" : (representante_L3, autotopismos_L3[2]),
    "L3_3" : (representante_L3, autotopismos_L3[3]),
    "L41_1" : (representante_L41, autotopismos_L41[1]),
    "L41_2" : (representante_L41, autotopismos_L41[2]),
    "L41_3" : (representante_L41, autotopismos_L41[3]),
    "L41_4" : (representante_L41, autotopismos_L41[4]),
    "L41_5" : (representante_L41, autotopismos_L41[5]),
    "L42_1" : (representante_L42, autotopismos_L42[1]),
    "L42_2" : (representante_L42, autotopismos_L42[2]),
    "L42_3" : (representante_L42, autotopismos_L42[3]),
    "L42_4" : (representante_L42, autotopismos_L42[4]),
    "L42_5" : (representante_L42, autotopismos_L42[5]),
    "L51_1" : (representante_L51, autotopismos_L51[1]),
    "L51_2" : (representante_L51, autotopismos_L51[2]),
    "L51_3" : (representante_L51, autotopismos_L51[3]),
    "L51_4" : (representante_L51, autotopismos_L51[4]),
    "L52_1" : (representante_L52, autotopismos_L52[1]),
    "L52_2" : (representante_L52, autotopismos_L52[2]),
}
```

```python
######################
# Variables Globales #
######################

numeroDeParticipantes = -1

def comparticionSecretosMononivel(numeroParticipantes):
    
    #########################
    # Variables principales #
    #########################
    
    representantesPosibles = list() #Almacenará todos los representantes para los que podamos obtener un conjunto critico
                                    #del tamaño deseado
        
    representantes = list() #Almacenará los representantes mayores entre los que se escogerá aleatoriamente alguno
        
    tamaños = list() #Almacenará el tamaño de cada uno de los posibles representantes
    
    ###############
    # Operaciones #
    ###############
    
    # 1) El primer paso consistirá en tomar el mayor representante posible (+ seguridad) y el autotopismo asociado
    
    for clave in comparteSecretos:
        #Comprobamos que el número de participantes esté dentro del rango scs y lcs
        if(numeroParticipantes>=comparteSecretos[clave][0] and numeroParticipantes<=comparteSecretos[clave][1]):
            representante = relacion[clave]
            representantesPosibles.append(representante)
            tamaños.append(len(representante[0]))
            
    #Una vez incluidos todos los posibles representantes, elegimos el mayor tamaño
    mayorTam = max(tamaños)
    
    #Recorremos la lista de representantes para seleccionar aquellos cuyo tamaño coincida con el máximo
    for repre in representantesPosibles:
        if(len(repre[0])==mayorTam):
            representantes.append(repre)
            
    #Seleccionamos uno aleatoriamente
    posAleatoria = random.randint(0, len(representantes)-1)
    
    #El par Representante - Autotopismo seleccionado será:
    representanteSeleccionado = representantes[posAleatoria]
    
    cuadrado = representanteSeleccionado[0] 
    print("El cuadrado representante elegido es:")
    pintaCuadrado(cuadrado)
    print("")
    autotopismo = representanteSeleccionado[1]
    print("El autotopismo elegido es: ", autotopismo)
    
    
    # 2) Una vez tenemos el representante y el autotopismo correspondiente, debemos buscar un θ conjunto crítico aleatorio del
    #    tamano del numero de participantes
            
    #Haremos combinaciones de n (n=numeroParticipantes) elementos entre las entradas del cuadrado seleccionado
    #(similar al estudio estadístico) y comprobaremos aleatoriamente todas ellas hasta que obtengamos un conjunto crítico
    
    ordenCuadrado = int(math.sqrt(len(cuadrado)))
    
    orbitasCuadrado = hallaOrbitas(cuadrado, autotopismo) #Calculamos las órbitas del cuadrado
    combinacionesOrbitas = list(itertools.combinations(orbitasCuadrado, numeroParticipantes)) #Hacemos las combinaciones
    conjuntoCriticoEncontrado = False #Condición de ruptura para el bucle de búsqueda de conjunto crítico
    
    while( not conjuntoCriticoEncontrado):
        #Seleccionamos una combinación aleatoria de entradas
        combAleatoria = random.randint(0,len(combinacionesOrbitas)-1)
        combSeleccionada = combinacionesOrbitas[combAleatoria]
        
        #Formamos el conjunto a evaluar:
        conjunto = [a[0] for a in combSeleccionada]
        
        #Comprobamos si es conjuntoCrítico
        cuadradoParcial = []
        
        #Creamos un cuadrado vacío inicial cuyo tamaño sea idéntico al del representante pasado por parámetro
        for _ in range(len(cuadrado)):
            cuadradoParcial.append([])
        
        #Para cada elemento del conjunto seleccionado, calculamos su posición y lo incluimos en el cuadrado parcial
        for j in conjunto:
            posicionEnCuadrado = j[0]*ordenCuadrado + j[1] 
            cuadradoParcial[posicionEnCuadrado] = j
        
        #Comprobamos si es únicamente completable
        cuadradoAux, res = esUnicamenteCompletable(cuadradoParcial,autotopismo)
        
        #Si es unicamente completable comprobaremos todos las combinaciones de n-1 conjuntos que se puedan
        #formar con sus entradas
        
        if(res):            
            #Obtenemos todas las combinaciones de n-1 elementos para el conjunto inicial (suponiendo n el tamaño del conjunto inicial)
            tamSubconjunto = len(conjunto)-1
            subconjuntos = list(itertools.combinations(conjunto,tamSubconjunto))
            unicamenteCompletableSubconjunto = False
            
            #Hacemos la comprobación con cada subconjunto
            for subconjunto in subconjuntos:

                res2 = compruebaSubconjunto(ordenCuadrado,subconjunto,autotopismo)                
                unicamenteCompletableSubconjunto = unicamenteCompletableSubconjunto or res2
            
                if(res2):                    
                    break #Impedimos que se ejecute más
            
            if(not unicamenteCompletableSubconjunto):
                conjuntoCriticoEncontrado = True    
    print("El conjunto critico inicial es: ", conjunto) 
    print("La combinación de órbitas seleccionada es: ", combSeleccionada)
    # 3) Una vez tenemos el conjunto crítico del tamaño deseado, generamos aleatoriamente un isotopismo y lo aplicamos al
    #    cuadrado que tenemos
    isotopismoAleatorio = generaIsotopismoAleatorio(ordenCuadrado)
    
    #Para mayor seguridad, evitaremos que se generen isotopismos triviales
    while(isotopismoAleatorio[0]==[] or isotopismoAleatorio[1]==[] or isotopismoAleatorio[2] == []):
        isotopismoAleatorio = generaIsotopismoAleatorio(ordenCuadrado)
    
    print("El isotopismo aleatorio es: ", isotopismoAleatorio)
    
    #Cuando tengamos el isotopismoAleatorio lo aplicamos al cuadrado. Esta será la clave del esquema de compartición.
    cuadradoClave = aplicaIsotopismoAleatorio(cuadrado,isotopismoAleatorio) 
    
    
    # 4) Calculamos el autotopismo transformado
    autotopismoTransf = autotopismoTransformado(autotopismo, isotopismoAleatorio, ordenCuadrado)
    
    # 5) Por último, recuperamos el conjunto crítico a partir del autotopismo transformado y el conjunto critico inicial
    
    conjuntoCriticoTransformado = []
    for ent in conjunto:
        entAux = aplicaIsotopismo(ent, isotopismoAleatorio)
        conjuntoCriticoTransformado.append(entAux)
        
    
    return [conjuntoCriticoTransformado, cuadradoClave, autotopismoTransf]
    
```
